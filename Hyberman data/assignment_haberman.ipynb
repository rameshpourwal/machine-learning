```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
import numpy as np

#load the data from datset 
data=pd.read_csv('haberman.csv')

#get the columns name of the data set

print(data.columns)
```

```python
#print the data of each class
print(data['status'].value_counts())
```

```python
# It is imbalanced data set because of the both the classes have different no of data
print(data.describe())

#data=data.drop(['status'],axis = 1, inplace = True)
```

```python
#pair data plot
sbn.pairplot(data,vars=['age','year','nodes'],hue='status',height=3).add_legend()


plt.suptitle('Pair Plot', y=1.08)
plt.show()
```

Conclusion:
1. From the above pair plot we can see that graph b/w age v/s nodes is some clearance.
2. Data is the imbalanced type.
3. People whome age is b/w 20 to 40 and node value is between 0 to 15 are survied 5 year or more

```python

sbn.set_style('whitegrid')
sbn.FacetGrid(data,hue='status',height=4).map(sbn.distplot,'nodes').add_legend()
plt.title("Node PDF Plot")
plt.show()
```

Conclusion From nodes PDF
1. pateinets whose node is 0 or 1 are more survided.
2. there are few chances to survived for node> 25.

```python
sbn.set_style('whitegrid')
sbn.FacetGrid(data,hue='status',height=4).map(sbn.distplot,'age').add_legend()
plt.title("Age PDF Plot")
plt.show()
```

Conclusion From Age PDF
1. major overlapping is observed , which tells as the patients age is irrespectivly for survied.
2. We can't decide the survive chance by age.

```python
sbn.set_style('whitegrid')
sbn.FacetGrid(data,hue='status',height=4).map(sbn.distplot,'year').add_legend()
plt.title("Year PDF Plot")
plt.show()
```

Conclusion From Year PDF
1. major overlapping is observed , which tells as the patients year is irrespectivly for survied.
2. We can't decide the survive chance by year.
3. We can see only that how many operation were successfully.
